```python
import pandas as pd
npr = pd.read_csv(r"C:\Users\akhil\OneDrive\Desktop\Natural-Language-Processing-with-Python-master\05-Topic-Modeling\quora_questions.csv")
display(npr)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_df = 0.95, min_df = 2, stop_words = "english")
dtm = tfidf.fit_transform(npr["Question"])
dtm
```

```python
#tfidf.get_feature_names()[3000]
from sklearn.decomposition import NMF
nmf_model = NMF(n_components=20, random_state=42)
nmf_model.fit(dtm)
```

```python
for i, topic in enumerate(nmf_model.components_):
    print(f"Top words from topic {i} are")
    for j in topic.argsort()[-15:]:
        print(tfidf.get_feature_names()[j], end=", ")
    print("\n")
    
```

```python
topic_result = nmf_model.transform(dtm)
npr["Topic"]=topic_result.argmax(axis=1)
npr
```
